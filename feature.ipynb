# Benchmark

```python
# ===== FIX PATH =====
sys.path.append(os.getcwd())

os.makedirs("output/benchmark_result", exist_ok=True)

# ===== LOOP INPUT =====
for i in range(1, 11):
    input_file = f"input/input_{i:02}.txt"

    if not os.path.exists(input_file):
        print(f"Skip {input_file}")
        continue

    print(f"\n=== Running {input_file} ===")

    # ===== CHẠY BENCHMARK =====
    df = benchmark_one(input_file)

    # ===== FILE RIÊNG =====
    output_file = f"output/benchmark_result/benchmark_{i:02}.csv"

    df.to_csv(output_file, index=False)

    print(f"✅ Saved {input_file} → {output_file}")
```
