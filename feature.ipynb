```python
import sys
import os
import pandas as pd

# ===== FIX PATH =====
sys.path.append(os.getcwd())

OUTPUT_PATH = "output/benchmark_result/benchmark.csv"
os.makedirs("output/benchmark_result", exist_ok=True)

# ===== LOOP INPUT =====
for i in range(1, 11):
    input_file = f"input/input_{i:02}.txt"

    if not os.path.exists(input_file):
        print(f"Skip {input_file}")
        continue

    print(f"\n=== Running {input_file} ===")

    # ===== CHẠY BENCHMARK =====
    df = benchmark_one(input_file)

    # ===== GHI NGAY VÀO CSV =====
    if os.path.exists(OUTPUT_PATH):
        df.to_csv(OUTPUT_PATH, mode='a', header=False, index=False)
    else:
        df.to_csv(OUTPUT_PATH, index=False)

    print(f"✅ Saved {input_file} to CSV")
```
